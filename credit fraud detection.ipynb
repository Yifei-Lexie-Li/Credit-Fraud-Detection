It is highly imbalanced, because the occurrences of fraud are rare compared to normal transactions.
Implication: Most machine learning models aim to minimize the overall error. In an imbalanced dataset, the model may predict all transactions as non-fraudulent to achieve high accuracy.

```python
#handling imbalanced dataset---downsampling
import sklearn
import imblearn
from imblearn.under_sampling import RandomUnderSampler
# Initialize RandomUnderSampler
rus = RandomUnderSampler(random_state=24)
```

```python
# Separate features (X) and target (y)
x = credit_df.drop('Class', axis=1)
y = credit_df['Class']
# Fit and apply the resampler to the data
x1, y1= rus.fit_resample(x, y)

# Convert the resampled data back to a DataFrame
down_df = pd.concat([pd.DataFrame(x1, columns=x.columns), pd.DataFrame(y1, columns=['Class'])], axis=1)

down_df.head()
```

```python
down_df.describe()
```

```python
down_df.shape
```

```python
sns.countplot(x='Class', data=down_df)
```

```python
#feature rescaling
from sklearn.preprocessing import RobustScaler, StandardScaler
standard_scaler = StandardScaler()
```

Since most of our data has already been scaled except Amount and Time.
We choose these 2 columns to standardize.

```python
modify_features=['Time','Amount']
down_df[modify_features] = standard_scaler.fit_transform(down_df[modify_features])
```

```python
down_df.describe()
```

```python
#Model Building
#split the data(80:20)
x2 = down_df.drop(['Class','Time'], axis=1)
y2 = down_df['Class']

#split the dataset
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x2, y2, test_size=0.2, random_state=24)
print(train_x)
```

```python
# Logistic Regression model
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, roc_curve
```

```python
model_strong = LogisticRegression(C=0.0001, solver='liblinear')
model_strong.fit(train_x, train_y)
pred_y_strong = model_strong.predict(test_x)
pred_prob_strong = model_strong.predict_proba(test_x)[:, 1]
```

```python
model_weak = LogisticRegression(C=100000000000, solver='liblinear')
model_weak.fit(train_x, train_y)
pred_y_weak = model_weak.predict(test_x)
pred_prob_weak = model_weak.predict_proba(test_x)[:, 1]
```

```python
#precision
precision_strong = precision_score(test_y, pred_y_strong)
print("Precision_strong:", precision_strong)
precision_weak = precision_score(test_y, pred_y_weak)
print("Precision_weak:", precision_weak)
```

```python
labels = ['Strong Model', 'Weak Model']
precisions = [precision_strong, precision_weak]
plt.bar(labels, precisions, color=['blue', 'orange'])
plt.xlabel('Model')
plt.ylabel('Precision')
plt.title('Precision Comparison between Strong and Weak Models')
plt.ylim(0, 1) 
plt.show()
```

```python
#recall
recall_strong = recall_score(test_y, pred_y_strong)
print("recall_strong:", recall_strong)
recall_weak = recall_score(test_y, pred_y_weak)
print("recall_weak:", recall_weak)
```

```python
labels = ['Strong Model', 'Weak Model']
recalls = [recall_strong, recall_weak]
plt.bar(labels, recalls, color=['blue', 'orange'])
plt.xlabel('Model')
plt.ylabel('Recall')
plt.title('Recall Comparison between Strong and Weak Models')
plt.ylim(0, 1)  
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, auc

# Calculate precision-recall curve for strong predictions
precision_strong, recall_strong, _ = precision_recall_curve(test_y, pred_y_strong)
auc_strong = auc(recall_strong, precision_strong)

# Calculate precision-recall curve for weak predictions
precision_weak, recall_weak, _ = precision_recall_curve(test_y, pred_y_weak)
auc_weak = auc(recall_weak, precision_weak)

# Plotting the Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall_strong, precision_strong, label=f'Strong (AUC = {auc_strong:.2f})', color='blue')
plt.plot(recall_weak, precision_weak, label=f'Weak (AUC = {auc_weak:.2f})', color='green')
plt.xlabel('Recall', fontsize=14)
plt.ylabel('Precision', fontsize=14)
plt.title('Precision-Recall Curve', fontsize=16)
plt.legend(loc='best', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

fpr_weak, tpr_weak, _ = roc_curve(test_y, pred_prob_weak)
auc_weak = auc(fpr_weak, tpr_weak)

fpr_strong, tpr_strong, _ = roc_curve(test_y, pred_prob_strong)
auc_strong = auc(fpr_strong, tpr_strong)

plt.figure(figsize=(8, 6))
plt.plot(fpr_weak, tpr_weak, label=f"Weak Model (AUC = {auc_weak:.2f})", color="green")


plt.plot(fpr_strong, tpr_strong, label=f"Strong Model (AUC = {auc_strong:.2f})", color="blue")
plt.plot([0, 1], [0, 1], 'k--', label="Random Model (AUC = 0.50)", color="green")

plt.title("ROC Curve for Strong and Weak Models")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
#f1_score
f1_score_strong = f1_score(test_y, pred_y_strong)
print("f1_score_strong:", f1_score_strong)
f1_score_weak = f1_score(test_y, pred_y_weak)
print("f1_score_weak:", f1_score_weak)
```

```python
labels = ['Strong Model', 'Weak Model']
f1_scores = [f1_score_strong, f1_score_weak]
plt.bar(labels, f1_scores, color=['blue', 'orange'])
plt.xlabel('Model')
plt.ylabel('F1 Score')
plt.title('F1 Score Comparison between Strong and Weak Models')
plt.ylim(0, 1)
plt.show()
```

```python
#roc_auc
roc_auc_score_strong = roc_auc_score(test_y, pred_prob_strong)
print("roc_auc_score_strong:", roc_auc_score_strong)
roc_auc_score_weak = roc_auc_score(test_y, pred_prob_weak)
print("roc_auc_score_weak:", roc_auc_score_weak)
```

```python
labels = ['Strong Model', 'Weak Model']
roc_auc_scores = [roc_auc_score_strong, roc_auc_score_weak]
plt.bar(labels, roc_auc_scores, color=['blue', 'orange'])
plt.xlabel('Model')
plt.ylabel('ROC-AUC Score')
plt.title('ROC-AUC Score Comparison between Strong and Weak Models')
plt.ylim(0, 1)  
plt.show()
```

Regularization: Since the score of weak model is usually larger, so regularization is not helpful in this case.

Why not use accuracy: In imbalanced datasets, even if the model predicts all samples as the majority class, it can still achieve a high accuracy. However, in this case, high accuracy does not indicate good model performance because it fails to effectively identify the minority class samples. Additionally, when we perform undersampling, it may cause the model to lose the ability to learn from the majority class samples, potentially discarding some useful information for the model, leading to lower accuracy.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_y, pred_y_weak)
TN, FP, FN, TP = cm.ravel()

print("True Negative (TN):", TN)
print("False Positive (FP):", FP)
print("False Negative (FN):", FN)
print("True Positive (TP):", TP)
```

```python
loss=500*FN+10*FP
print('Financial Loss is:',loss)
```

optimize the loss: Adjust the threshold to minimize both false negatives (FN) and false positives (FP) as much as possible.

```python
threshold = 0.3  
pred_y_best= (pred_prob_weak >= threshold).astype(int)
```

```python
cm = confusion_matrix(test_y, pred_y_best)
TN, FP, FN, TP = cm.ravel()

print("True Negative (TN):", TN)
print("False Positive (FP):", FP)
print("False Negative (FN):", FN)
print("True Positive (TP):", TP)
```
